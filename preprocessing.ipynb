```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def roberts_cross(img):
    kernelx = np.array([[1, 0], [0, -1]])
    kernely = np.array([[0, 1], [-1, 0]])
    x = cv2.filter2D(img, -1, kernelx)
    y = cv2.filter2D(img, -1, kernely)
    img = x ** 2 + y ** 2
    img = np.sqrt(img)
    img = img.astype(np.float64)
    return img
```

```python
img = cv2.imread("ZhangLabData/CellData/chest_xray/train/PNEUMONIA/BACTERIA-1008087-0001.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img = cv2.blur(img, (3,3))
plt.imshow(img, cmap='gray')
```

```python
img = roberts_cross(img)
plt.imshow(img, cmap='gray')
```

```python
img2 = cv2.imread("ZhangLabData/CellData/chest_xray/train/NORMAL/NORMAL-1003233-0001.jpeg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(img2, cmap='gray')
```
